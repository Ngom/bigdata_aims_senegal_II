```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
brd = load_iris()
```

```python
data = brd.data
```

```python
plt.scatter(data[:,1],data[:,2],cmap='brg')
```

```python
################# KMean ############################################
```

```python
from sklearn.cluster import KMeans
kms  = KMeans(n_clusters=2)
kms.fit(data[:,1:3:1])
y_pred = kms.predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred,cmap='brg')
```

```python
###########################  Meanshift ##########################################
```

```python
from sklearn.cluster import MeanShift
msh = MeanShift(bandwidth=1.0)
```

```python
y_predi = msh.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_predi,cmap='brg')
```

```python
########################## Affinity Propagation ###################################""
```

```python
from sklearn.cluster import AffinityPropagation 
afp = AffinityPropagation(damping=0.9) # damping impacts on 
y_pred_1 = afp.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred_1,cmap='brg')
```

```python
np.unique(afp.labels_).size # pou eviter les doublons
```

```python
np.shape(afp.labels_)
```

```python
############################## Agglomerative Clustering ####################################
```

```python
from sklearn.cluster import AgglomerativeClustering
alc = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='average')
y_pred_2 = alc.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred_2,cmap='brg')
```

```python
############################## DBSCAN #####################################""
```

```python
from sklearn.cluster import  DBSCAN
dbscan =  DBSCAN(eps=0.5,min_samples=3)
y_pred_3 = dbscan.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred_3,cmap='brg')
```
