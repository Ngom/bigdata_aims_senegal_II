```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.normal(0.0,3.5,501)
y = np.random.normal(0.0,0.5,501)
theta = -np.pi/6.0
x1 = x*np.cos(theta) + y*np.sin(theta)
y1 = y*np.cos(theta) - x*np.sin(theta)
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x1,y1,cmap='brg')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(np.c_[x1,y1])
```

```python
vector = pca.components_
vector
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x1,y1,cmap='brg')
plt.quiver(np.mean(x1),np.mean(y1),vector[1][1],vector[0][1],angles='xy',scale=5.0,color='r')
plt.quiver(np.mean(x1),np.mean(y1),vector[0][1],vector[1][0],angles='xy',scale=5.0,color='r')
```

```python
pca.explained_variance_ratio_
```

```python
pca.get_covariance()
```
