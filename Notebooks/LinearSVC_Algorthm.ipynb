```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = iris.data[(iris.target==0)|(iris.target==1),0:3:2]
np.shape(X)
```

```python
y = iris.target[(iris.target==0)|(iris.target==1)]
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import LinearSVC
```

```python
lsvc = LinearSVC(dual=False,C=0.1,petality='l1',tol=)
lsvc.fit(X,y)
```

```python
xx,yy = np.meshgrid(np.linspace(0.0,10.0,201), np.linspace(0.0,10.0,201))
pred = lsvc.predict(np.c_[xx.ravel(),yy.ravel()])
Z = pred.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(16,7))
plt.contourf(xx,yy,Z,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=y)
plt.axis('scaled')
plt.xlabel('Blue in high level')
plt.ylabel('Red in in slow level')
plt.title('SVC',fontsize=25)
plt.xlim(2.0,10.0)
plt.ylim(0.0,8.0)
```
