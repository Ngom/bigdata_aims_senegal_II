```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = iris.data[(iris.target==0)|(iris.target==1),1:3:2]
np.shape(X)
```

```python
y = iris.target[(iris.target==0)|(iris.target==1)]
```

```python
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import LinearSVC
```

```python
lsvc = LinearSVC(C=0.01)
lsvc.fit(X,y)
```

```python
#plt.scatter(X.T[0],X.T[1],c=y,cmap='spetral_r')
```

```python
xx,yy = np.meshgrid(np.linspace(0.0,3.0,201), np.linspace(-2.0,2.0,201))
pred = lsvc.predict(np.c_[xx.ravel(),yy.ravel()])
Z = pred.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.contourf(xx,yy,Z,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=y)
plt.axis('scaled')
plt.xlabel('Blue in high level')
plt.ylabel('Red in in slow level')
plt.title('SVC',fontsize=20)
plt.xlim(0.0,3.0)
plt.ylim(-2.0,2.0)
```

```python
plt.scatter(iris.data[:,0],iris.data[:,3])
```

```python
x1 = iris.data[(iris.target==0)|(iris.target==1),1:4:2]
np.shape(x1)
x2 = iris.target[(iris.target==0)|(iris.target==1)]
x1 = scale(x1)
```

```python
lsvc1 = LinearSVC(C=0.01)
lsvc1.fit(x1,x2)
```

```python
xx,yy = np.meshgrid(np.linspace(-4.0,4.0,201), np.linspace(-3.0,3.0,201))
pred1 = lsvc1.predict(np.c_[xx.ravel(),yy.ravel()])
Z1 = pred1.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.contourf(xx,yy,Z1,alpha=0.4)
plt.scatter(x1.T[0],x1.T[1],c=x2)
plt.axis('scaled')
plt.xlabel('Blue in high level')
plt.ylabel('Red in in slow level')
plt.title('SVC',fontsize=20)
plt.xlim(-4.0,4.0)
plt.ylim(-3.0,3.0)
```

```python
lsvc1.score(x1,x2)
```
